```python
import pandas as pd 
df = pd.read_csv('PRPL-Facebook-Government-Report-2022_H1.csv')
df
```

```python
df.replace(',','', regex=True, inplace=True) 
df
```

```python
df['Total Requests'] = df['Total Requests'].astype(int)
```

```python
df = df.sort_values( by = 'Total Requests' , ascending = False).head(10)
df
```
